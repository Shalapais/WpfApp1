```python
import matplotlib.pyplot as plt
import numpy as np

# Данные для шести типов экранов (по шкале от 0 до 3, где 0 – плохо/нет, 3 – отлично/да)
# Критерии: Мультитач, Реакция на любой предмет, Морозостойкость, Вандалоустойчивость, Светопропускание
data = {
    'Резистивные':        [0, 3, 0, 1, 1],
    'Емкостные':          [0, 0, 2, 2, 3],
    'Проекционно-емкостные': [3, 0, 3, 3, 3],
    'Матричные':          [3, 3, 0, 1, 1],
    'Инфракрасные':       [0, 3, 3, 2, 3],
    'Тензометрические':   [0, 2, 1, 3, 3]
}

categories = ['Мультитач', 'Любой предмет', 'Морозостойкость', 'Вандалоустойчивость', 'Светопропускание']
types = list(data.keys())
values = np.array(list(data.values()))

# Настройка позиций для столбцов
x = np.arange(len(categories))  # позиции для групп критериев
width = 0.12  # ширина одного столбца

fig, ax = plt.subplots(figsize=(12, 6))

# Рисуем столбцы для каждого типа экрана
for i, (t, v) in enumerate(data.items()):
    ax.bar(x + i * width, v, width, label=t)

# Добавляем подписи
ax.set_xlabel('Критерии')
ax.set_ylabel('Оценка (0-3)')
ax.set_title('Сравнение характеристик типов сенсорных экранов')
ax.set_xticks(x + width * (len(types) - 1) / 2)
ax.set_xticklabels(categories)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax.set_ylim(0, 3.5)

# Добавляем сетку для удобства чтения
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd


df = pd.read_csv('Ice_cream selling data.csv')
df.head()
```

```python
from matplotlib import pyplot as plt

x = df['Temperature (°C)']
y = df['Ice Cream Sales (units)']


plt.plot(x,y,'b+')
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales (units)')
plt.title('Фигуративное облако точек')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
#help(train_test_split)

X_train, X_test, y_trin, y_test = train_test_split(x,y, shuffle=True, test_size=0.2, random_state = 42)

print('Количестро записей в наборе для обучение',len(X_train))
print('Количестро записей в наборе для тестирования',len(X_test))

print(X_train)
print(type(X_train))
```

```python
import numpy as np

info = """
    Для моделирования используем функцию polyfit
    из библиотеки numpy
"""

coeff = np.polyfit(X_train, y_trin,deg = 2)

print(coeff)

#Модель
f_x = lambda x: coeff[0]*x**2 + coeff[1]*x + coeff[2]

X_model = list(range(-5,6))
y_model = [f_x(x) for x in X_model]

plt.plot(X_model,y_model,'r-')
plt.plot(X_test,y_test,'b+')
plt.title('Полином 2 порядка и данные для тестирования')
plt.show()
```
